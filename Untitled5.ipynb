```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# دانلود مجموعه داده‌های مورد نیاز
nltk.download('vader_lexicon')

# متن مورد تحلیل
text = " I do realy like this movie it was so nice"
# تحلیل احساسات
sia = SentimentIntensityAnalyzer()
sentiment_scores = sia.polarity_scores(text)

# نمایش نتایج
print(sentiment_scores)

# نمودار احساسات
plt.bar(['Negative', 'Neutral', 'Positive', 'Compound'],
        [sentiment_scores['neg'], sentiment_scores['neu'], sentiment_scores['pos'], sentiment_scores['compound']])
plt.title('Sentiment Analysis')
plt.xlabel('Sentiment')
plt.ylabel('Score')
plt.show()
```
